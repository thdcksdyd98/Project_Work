```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.firefox.options import Options
import pandas as pd
import time
```

```python
options = Options()
# options.add_argument("start-maximized")
# options.add_argument("disable-infobars")
options.add_argument("--disable-extensions")
options.add_argument('--no-sandbox')
options.add_argument('--disable-application-cache')
options.add_argument('--disable-gpu')
options.add_argument("--disable-dev-shm-usage")
```

```python
# dr = webdriver.Firefox("C:/windows/chromedriver.exe")
dr = webdriver.Firefox(options = options)
dr.maximize_window
dr.get('https://www.sephora.com/happening/storelist')
time.sleep(5)
```

```python
columns = ['state','address']
result_df = pd.DataFrame(columns=columns)
```

```python
result_list =  []

for idx in range(4,100, 2):
    counter = 1
    time.sleep(4)
    for _ in range(len(dr.find_elements(By.XPATH, f'/html/body/div[2]/div[2]/main/div[{idx}]//li'))):

        tmp = []

        state = dr.find_element(By.XPATH, f'/html/body/div[2]/div[2]/main/div[{idx}]/h3').get_attribute('innerHTML')
        tmp.append(state)
        
        try:
            
            clickable = dr.find_element(By.XPATH, f'/html/body/div[2]/div[2]/main/div[{idx}]/ul/li[{counter}]/a')
            
            if ', ' == clickable.get_attribute('innerHTML'):
                continue

            time.sleep(5)
            clickable.click()
            time.sleep(5)

            try:
                if dr.find_element(By.XPATH, "//*[@id='modal0Dialog']"):
                    exit = dr.find_element(By.XPATH, '//*[@id="modal0Dialog"]/button')
                    exit.click()
            except:
                pass

            store_info = dr.find_element(By.XPATH, '//*[@data-at="store_location"]')
            add = store_info.text.replace('\n',' ').split(' US ')[0]
            print(add)
            tmp.append(add)
        
        except:
            pass
        
        counter += 1
        dr.back()

        result_list.append(tmp)
```

```python
test = pd.DataFrame(result_list)
test.to_csv("PATH")
```

```python
new_df = pd.DataFrame(result_list, columns=columns)
df = pd.concat([result_df, new_df], ignore_index=True)
```
